# Model Training

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
# separe with the target variable
features = df[['SepalLengthCm','SepalWidthCm','PetalLengthCm','PetalWidthCm']]
y = df['Species']
```

```python
# encoding the target variable
label = LabelEncoder()
y = label.fit_transform(y)
```

```python
# split our data in the training and testing set
# split our data in the training and testing set
X_train,X_test,y_train,y_test=train_test_split(features,y,test_size=0.3,random_state=100)
```

```python
# Applying decision tree algorithm
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score,confusion_matrix
classifier= DecisionTreeClassifier(random_state=100)
#Fitting the training data
classifier.fit(X_train,y_train)
```

For example if in training data features value[1.6,5.6,3.2,1.0] belongs to category 2 (virginica),
means when we entered similar values or close values to model to predict category of fleur it wil
return 2. 

```python
# Prediction on random data for this example 
X=[[1.6,5.6,3.2,1.0]]
y_pred = classifier.predict(X)
y_pred
```

```python
# Prediction on X_test ( testing data)
y_pred = classifier.predict(X_test)
y_pred
```

```python
# create dataframe pour our values predict
prediction = pd.DataFrame({'Actual':y_test.squeeze(),'Prediction':y_pred.squeeze()})
prediction
```

Accuray of model

```python
# let's take a look how our model is performing
acc = accuracy_score(y_test,y_pred)
print("accuracy score is:",acc)
```

The accuracy of our model is quite good so it works well.We get accuracy 95% which is good, Means our
model pedict 5% wrong prediction about a category of flower.

# Model evaluation 

```python
mtx = confusion_matrix(y_test,y_pred)
sns.heatmap(mtx,annot=True,cmap='Blues')
plt.title("confusion matrix")
```

Confusion matrix is also good way to see model's performance. Here out of 150 samples 30%=45 samples
we used for testing purpose.
In confusion matrix the positions:
(0,0)-16 means 16 flowers of category 0 predicted right
(1,1)-10 flowers of category 1 predicted right
(2,2)-17 flowers of category 2 predicted right
(0,0),(1,1),(2,2) correct prediction rest wrong prediction
conclusion: diagonal positions of matrix-right prediction ,rest-wrong

```python
# Tree plotting/ decision making in decision tree
from sklearn import tree
featu_nm =['sepalLengthCm','sepalWidthCm','petalLengthCm','petalWidthCm']
class_nm =['Iris-setosa','Iris-verticolor','Iris-virginica']

fig = plt.figure(figsize=(25,20))
_=tree.plot_tree(classifier,feature_names=featu_nm,class_names=class_nm,filled=True)
plt.title("Decision Tree visualization",fontsize=24)
```
